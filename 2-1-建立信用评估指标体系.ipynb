```python
# 导入第 1 部分的阶段性数据
data = pd.read_csv(PREFIX + 'data-1.csv')
data.describe()
```

```python
def fill_div_na_inf(series):
    series.replace(np.inf, np.nan, inplace=True)
    series.fillna(0, inplace=True)
    return series
```

```python
# 年消费总额指标
data['trans_total'] = data.transCnt_mean * data.transAmt_mean

# 年取现总额指标
data['total_withdraw'] = data.cashCnt_mean * data.cashAmt_mean

# 平均每笔取现金额指标
data['avg_per_withdraw'] = fill_div_na_inf(data.cashTotalAmt / data.cashTotalCnt)

# 网上平均每笔消费额指标
data['avg_per_online_spend'] = fill_div_na_inf(data.onlineTransAmt / data.onlineTransCnt)

# 公共事业平均每笔缴费额指标
data['avg_per_public_spend'] = fill_div_na_inf(data.publicPayAmt / data.publicPayCnt)

# 不良记录指标
data['bad_record'] = data.inCourt + data.isDue + data.isCrime + data.isBlackList

# 只保留 6 位小数
data = data.round(6)
```

```python
data.to_csv(PREFIX + 'data-2.csv', index=False)
```
